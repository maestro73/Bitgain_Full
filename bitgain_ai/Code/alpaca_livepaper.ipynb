```python
import alpaca_trade_api as tradeapi
import time
import datetime
from datetime import timedelta
from pytz import timezone
import pandas as pd
import talib
import numpy as np
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
#from sklearn.metrics import scorer
from sklearn.metrics import accuracy_score

import requests
import bs4 as bs

tz = timezone('America/New_York')

api = tradeapi.REST('PKHK6KFHV5Q1YGG5850W',
                    '2sXy3eai4LwvLcbkFVN9gj1HojtH6MEO3/EkmPFw',
                    'https://paper-api.alpaca.markets')

import logging
logging.basicConfig(filename='./new_sklearn_SVC_15min.log', format='%(name)s - %(levelname)s - %(message)s')
logging.warning('{} logging started'.format(datetime.datetime.now().strftime("%x %X")))

stocks = ['ABC','AEE','AFL','AKAM','ALXN','AMG','AMT','ANET','ANTM','ARE','AVB','BBY','BF.B','BLL','BMY','CAH','CERN','CHTR',
          'CINF','CLX','CMG','COG','COO','COP','COTY','CTAS','CXO','DLTR','DRI','DTE','DVN','DWDP','EA','ED','EL','ESS',
          'ETR','EVRG','EXPE','EXR','FANG','FE','FL','FLT','GD','GE','GIS','GPN','HIG','HRL','HSY','INFO','INTU','IP','IQV',
          'JCI','JNJ','KHC','KMB','KMX','KO','KR','LB','LEN','LH','MCO','MDT','MKC','MNST','MO','MRK','MSCI','MTB','MTD','MYL',
          'NKE','OMC','PAYX','PEP','PFE','PHM','PM','PRGO','PSA','PSX','PXD','RE','RHI','RHT','RL','RMD','ROL','ROP','RSG',
          'SBUX','SEE','SIVB','SJM','SNA','SO','SRE','SYF','SYK','TAP','TGT','TIF','TRV','TTWO','UDR','UHS','VAR','VRSK',
          'VZ','WBA','WEC','WELL','WLTW','WM','XEC','XEL','XLNX','YUM','ZBH','ZTS']

def save_sp500_tickers():
    resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    soup = bs.BeautifulSoup(resp.text, 'lxml')
    table = soup.find('table', {'class': 'wikitable sortable'})
    tickers = []
    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[1].text #change to 0 to get company names instead of tickers
        tickers.append(ticker)
    #with open("sp500tickers.pickle","wb") as f:
        #pickle.dump(tickers,f)
    return sorted(tickers)

sp500 = save_sp500_tickers()

# if market is not open, calculates datetime.timedelta until market open
def time_to_open(current_time):
    if current_time.weekday() <= 4:
        d = (current_time + timedelta(days=1)).date()
    else:
        days_to_mon = 0 - current_time.weekday() + 7
        d = (current_time + timedelta(days=days_to_mon)).date()
    next_day = datetime.datetime.combine(d, datetime.time(9, 30, tzinfo=tz))
    seconds = (next_day - current_time).total_seconds()
    return seconds

# loops every 15 min with time.sleep(15*60)
while True:
    start_time = time.time()
    # Check if Monday-Friday
    if datetime.datetime.now(tz).weekday() >= 0 and datetime.datetime.now(tz).weekday() <= 4:
        # Checks if between 9:30 and 3:30
        print('Trading day', str(datetime.datetime.now(tz).time()))
        if datetime.datetime.now(tz).time() > datetime.time(9, 30) and datetime.datetime.now(tz).time() <= datetime.time(15, 30):
            # Market open confirmed!
            for ticker in sp500:
                try:
                    shares = round(1000/api.get_barset(ticker, '15Min', limit=1).df[ticker].close[0])
                    loop_start = time.time()
                    #download current ticker data
                    print(" ")
                    print("********************************")
                    print(ticker, " starting: ")
                    #Df = api.polygon.historic_agg('minute', ticker, limit=3000).df #.iloc[::4,:]
                    data = api.get_barset(ticker, '15Min', limit=1000).df.loc[:,ticker] # downloads 1000 ticks of "15min" data for current ticker
                    data.loc[:, 'open-close'] = data['open'] - data['close'] # custom open minus close feature
                    data.loc[:, 'high-low'] = data['high'] - data['low'] # custom high minus low feature
                    data.loc[:, 'MACD'] = talib.MACD(data['close'].values)[0]
                    data.loc[:, 'MACD_SIG'] = talib.MACD(data['close'].values)[1]
                    data.loc[:, 'MACD_HIST'] = talib.MACD(data['close'].values)[2]
                    data.loc[:, 'RSI'] = talib.RSI(data['close'].values)

                    X_temp = data.drop(columns=['open','high','low','volume']).dropna() # drops all rows with "NaN" and raw data columns
                    y = np.where(X_temp['close'].shift(-1) > X_temp['close'],1,-1) # y is 1 if price went up, -1 if down
                    X = X_temp.drop(columns='close') # finishes dropping raw data columns


                    # split data into training / test
                    #### change this here to maximize training set while allowing room in test set for indicator calculations and dropna(); should be >40 in test
                    split_percentage = 0.8
                    split = int(split_percentage*len(X))
                    # Train data set
                    X_train = X[:split].values
                    y_train = y[:split]
                    # Test data set
                    X_test = X[split:].values
                    y_test = y[split:]

                    # train
                    sc = StandardScaler()
                    X_train = sc.fit_transform(X_train)
                    X_test = sc.transform(X_test)
                    cls_svc = SVC().fit(X_train, y_train)
                    cls_knn = KNeighborsClassifier().fit(X_train, y_train)
                    
                    predictions_svc = cls_svc.predict(X_test)
                    predictions_knn = cls_knn.predict(X_test)
                    predicted_signal_svc = predictions_svc[-1] # prediction of current tick
                    predicted_signal_knn = predictions_knn[-1] # prediction of current tick
                    accuracy_svc = accuracy_score(y_test, predictions_svc)
                    accuracy_knn = accuracy_score(y_test, predictions_knn)
                    print('KNN Accuracy:{: .2f}%'.format(accuracy_knn*100))
                    print('KNN Prediction: ', predicted_signal_knn)
                    print('SVC Accuracy:{: .2f}%'.format(accuracy_svc*100))
                    print('SVC Prediction: ', predicted_signal_svc)
                    
                    if predicted_signal_svc == 1 and predicted_signal_knn == 1 and accuracy_svc >= 0.5 and accuracy_knn >= 0.5: # if both algo's agree on up and test accuracy >= 50%
                        if ticker not in [x.symbol for x in api.list_positions()]: # makes sure we don't already own it
                            logging.warning('{} {} - {}'.format(datetime.datetime.now(tz).strftime("%x %X"), ticker, shares))
                            api.submit_order(ticker, shares, 'buy', 'market', 'day') # buys shares
                            print(datetime.datetime.now(tz).strftime("%x %X"), 'buying', ticker)
                    else: # if one algo does not agree or test accuracy less than 50%
                        if ticker in [x.symbol for x in api.list_positions()]:  # makes sure we do own and available to sell
                            try:
                                api.submit_order(ticker, shares, 'sell', 'market', 'day') # sells shares
                                logging.warning('{} {} - {}'.format(datetime.datetime.now(tz).strftime("%x %X"), ticker, shares))
                                print(datetime.datetime.now(tz).strftime("%x %X"), 'selling', ticker)
                            except Exception as e:
                                print('Sell failed', ticker, e)

                    loop_end = time.time()
                    print("Loop run-time: ", (loop_end - loop_start))
                except Exception as e:
                    print("ERROR: ", e)
                    print(ticker, "FAILED!!!")
                    logging.warning(ticker, "logic FAILED!!!")
                    logging.warning("ERROR: ", e)
                    loop_end = time.time()
                    print("Loop run-time: ", (loop_end - loop_start))

        else:
            # Get time amount until open, sleep that amount
            print('Market closed ({})'.format(datetime.datetime.now(tz)))
            print('Sleeping', round(time_to_open(datetime.datetime.now(tz))/60/60, 2), 'hours')
            time.sleep(time_to_open(datetime.datetime.now(tz)))
    else:
        # If not trading day, find out how long until open, sleep that amount
        print('Market closed ({})'.format(datetime.datetime.now(tz)))
        print('Sleeping', round(time_to_open(datetime.datetime.now(tz))/60/60, 2), 'hours')
        time.sleep(time_to_open(datetime.datetime.now(tz)))

    end_time = time.time()
    sleep_length = (60*15-(end_time - start_time))
    print("Sleeping...", sleep_length, "seconds")
    time.sleep(sleep_length)


'''
def get_data_bars(symbols, rate, slow, fast):
    #data = api.get_barset(symbols, rate, limit=20).df
    data1 = api.get_barset(symbols[:100], rate, limit=20).df
    data2 = api.get_barset(symbols[100:200], rate, limit=20).df
    data3 = api.get_barset(symbols[200:300], rate, limit=20).df
    data4 = api.get_barset(symbols[300:400], rate, limit=20).df
    data5 = api.get_barset(symbols[400:500], rate, limit=20).df
    data6 = api.get_barset(symbols[500:], rate, limit=20).df
    data = pd.concat([data1, data2, data3, data4, data5, data6], axis=1)
    for x in symbols:
        data.loc[:, (x, 'fast_ema')] = data[x]['close'].rolling(window=fast).mean()
        data.loc[:, (x, 'slow_ema')] = data[x]['close'].rolling(window=slow).mean()
    return data

def get_signal_bars(symbol_list, rate, ema_slow, ema_fast):
    data = get_data_bars(symbol_list, rate, ema_slow, ema_fast)
    signals = {}
    for x in symbol_list:
        if data[x].iloc[-1]['fast_ema'] > data[x].iloc[-1]['slow_ema']: signal = 1
        else: signal = 0
        signals[x] = signal
    return signals

def time_to_open(current_time):
    if current_time.weekday() <= 4:
        d = (current_time + timedelta(days=1)).date()
    else:
        days_to_mon = 0 - current_time.weekday() + 7
        d = (current_time + timedelta(days=days_to_mon)).date()
    next_day = datetime.datetime.combine(d, datetime.time(9, 30, tzinfo=tz))
    seconds = (next_day - current_time).total_seconds()
    return seconds

def run_checker(stocklist):
    print('run_checker started')
    while True:
        # Check if Monday-Friday
        if datetime.datetime.now(tz).weekday() >= 0 and datetime.datetime.now(tz).weekday() <= 4:
            # Checks market is open
            print('Trading day', str(datetime.datetime.now(tz).time()))
            if datetime.datetime.now(tz).time() > datetime.time(9, 30) and datetime.datetime.now(tz).time() <= datetime.time(15, 30):
                signals = get_signal_bars(stocklist, '5Min', 20, 5)
                for signal in signals:
                    if signals[signal] == 1:
                        if signal not in [x.symbol for x in api.list_positions()]:
                            logging.warning('{} {} - {}'.format(datetime.datetime.now(tz).strftime("%x %X"), signal, signals[signal]))
                            api.submit_order(signal, 5, 'buy', 'market', 'day')
                            print(datetime.datetime.now(tz).strftime("%x %X"), 'buying', signal)
                    else:
                        if signal in [x.symbol for x in api.list_positions()]:
                            try:
                                api.submit_order(signal, 5, 'sell', 'market', 'day')
                                logging.warning('{} {} - {}'.format(datetime.datetime.now(tz).strftime("%x %X"), signal, signals[signal]))
                                print(datetime.datetime.now(tz).strftime("%x %X"), 'selling', signal)
                            except Exception as e:
                                print('Sell failed', signal, e)
                                #pass
                time.sleep(60)
            else:
                # Get time amount until open, sleep that amount
                print('Market closed ({})'.format(datetime.datetime.now(tz)))
                print('Sleeping', round(time_to_open(datetime.datetime.now(tz))/60/60, 2), 'hours')
                time.sleep(time_to_open(datetime.datetime.now(tz)))
        else:
            # If not trading day, find out how much until open, sleep that amount
            print('Market closed ({})'.format(datetime.datetime.now(tz)))
            print('Sleeping', round(time_to_open(datetime.datetime.now(tz))/60/60, 2), 'hours')
            time.sleep(time_to_open(datetime.datetime.now(tz)))



run_checker(stocks)
'''


```
