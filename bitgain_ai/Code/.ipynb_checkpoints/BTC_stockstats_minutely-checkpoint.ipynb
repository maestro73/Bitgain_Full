```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualization library  
import matplotlib.pyplot as plt
import time
from tqdm import tqdm
import numpy as np
import stockstats as ss

from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix,accuracy_score
from sklearn.feature_selection import SelectKBest,chi2,RFE,RFECV

import random
```

```python
stock = ss.StockDataFrame.retype(pd.read_csv('../Data/coinbaseUSD_1-min_data_2014-12-01_to_2018-11-11.csv'))
#stock = pd.read_csv("bitcoin_daily.csv")
#stock.dtypes
#stock["close"] = pd.to_numeric(stock["close"])
stock.dtypes
#stock.head()
#stock = ss.StockDataFrame.retype(stock)
type(stock)
```

```python
def profitable(x):
    if x["change"] > 0:
        return 1
    elif x["change"] <= 0:
        return 0
```

```python
stock = stock[1:].dropna() #removes all rows containing any NaN values
stock.head()
type(stock)
stock = ss.StockDataFrame.retype(stock)
type(stock)
stock
```

```python
stock['macd']
stock['volume_delta']
stock['open_-1_r']
stock['cr']
stock['cr-ma1']
stock['cr-ma2']
stock['cr-ma3']
stock["rsi_6"]
stock["rsi_12"]
stock["wr_10"]
stock["cci"]
stock['tr']
stock['atr']
stock['dma']
stock['pdi']
stock['mdi']
stock['dx']
stock['adx']
stock['adxr']
stock['trix']
stock['trix_9_sma']
stock['vr']
```

```python
stock.shape
```

```python
stock_clean = stock
stock = stock[len(stock)-44640:]
stock
```

```python
stock_clean["profitable"] = stock.apply(profitable, axis=1)
```

```python
stock_clean.head()
```

```python
stock_clean["profitable"] = stock_clean["profitable"].shift(-1) #shifts "profitable" target column up 1 so current days stats predict tomorrow's profitability 
```

```python
stock_clean.tail()
```

```python
stock_clean = stock_clean[1:len(stock_clean.index)-1] #removes last entry that now has NaN "profitable" column
```

```python
stock_clean
```

```python
stock_clean = stock_clean[1:].dropna() #removes all rows containing any NaN values
```

```python
stock_clean
```

```python
stock_clean
```

```python
y = stock_clean["profitable"]
x = stock_clean.drop(["profitable", "change"], axis=1)
x_std = (x - x.mean())/x.std()
```

```python
x_std = x_std.drop(["open","high","low","close","volume","timestamp"], axis=1)
```

```python
x_std
```

```python
import random
RANDOM_STATE = 123
random.seed(RANDOM_STATE)
x_train,x_test,y_train,y_test = train_test_split(x_std,y,test_size = .3,random_state = RANDOM_STATE)
```

```python
clf_rf = RandomForestClassifier(random_state = RANDOM_STATE)
clr_rf = clf_rf.fit(x_train,y_train)

predictions = clf_rf.predict(x_test)

ac = accuracy_score(predictions,y_test)
cm = confusion_matrix(predictions,y_test)

print('Accuracy Score: %f' % ac)
sns.heatmap(cm,annot = True,fmt = 'd')
```
