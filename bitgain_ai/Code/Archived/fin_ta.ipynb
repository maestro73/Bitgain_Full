```python
returns = returns.rename(columns={symbol: "returns"})
log_returns = log_returns.rename(columns={symbol: "log_returns"})
```

```python
df_w_returns = df.join(returns).join(log_returns)
```

```python
def profitable(x):
    if x["returns"] > 0:
        return 1
    elif x["returns"] <= 0:
        return 0
```

```python
df_w_returns["profitable"] = df_w_returns.apply(profitable, axis=1)
```

```python
df_profitable = df_w_returns.drop(["returns", "log_returns"], axis=1)
```
