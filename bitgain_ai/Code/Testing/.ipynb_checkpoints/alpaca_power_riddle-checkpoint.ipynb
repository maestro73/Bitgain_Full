```python
import alpaca_trade_api as tradeapi
from pytz import timezone
import pandas as pd

def is_prime(x):
    if x < 2:
        return False
    elif x == 2:
        return True  
    for n in range(2, x):
        if x % n ==0:
            return False
    return True
```

```python
S = api.polygon.historic_agg('day', 'SPY', limit=316).df.iloc[:-65,:]
D = api.polygon.historic_agg('day', 'DIA', limit=316).df.iloc[:-65,:]

#print(type(S))

S_dates = []
D_dates = []

for date, data in S.iterrows():
    if is_prime(int(data.close * 10**2)):
        #print("Prime****")
        S_dates.append(date)
        
for date, data in D.iterrows():
    if is_prime(int(data.close * 10**2)):
        #print("Prime****")
        D_dates.append(date)
        
#print(S_dates)
#print(D_dates)

prime_dates = (set(S_dates) & set(D_dates))
prime_dates_friday = []

#print( "ANSER" )
#print(prime_dates)

for date in prime_dates:
    if date.weekday() >= 3:
        prime_dates_friday.append(date)
        
print("   REAL ANSWER    ")
print(prime_dates_friday)

for x in range(len(prime_dates_friday)):
    print("Round X: ", x)
    print("Date: ", prime_dates_friday[x])
    print("S Price: ")
    print(S.loc[prime_dates_friday[x]])
    print("D Price: ")
    print(D.loc[prime_dates_friday[x]])
```
